## 3. Investment Strategies:
If we buy a \$1 milion house with a \$200,000 down payment then assuming a 6% ror on the house and a 2% interest rate, we'll be getting $\frac{1,000,000*(1+0.06-0.02)-1,000,000}{200,000} = 0.2$ per annum. Under an optimistic forecast of the housing sector, this would be the ideal strategy where you're making leveraged gains and saving money on housing. However, in this strategy we're overly exposed and sensitive to a downturn. Would it be better instead to buy a cheaper apartment and invest in shares instead?

The following is a discrete simulation of 5 different investment strategies under: Pessimistic, Medium and Optimistic performance of housing in a 13 year span. 

Seems like even under pessimistic forecasts, renting comes out bottom. Under Optimistic forecasts, one should invest in housing. Worst of all is to rent, which is essentially linear driven only by income. Finally, we see that a consistently reliable strategy is to purchase a cheap apartment to live in and invest the remaining into shares. 

```python
import dotmap
import logging
import pandas as pd

from args import *
from simulation import *

def run_simulation(ticks, args):
    values = []
    world = SimpleWorld(dotmap.DotMap(args))
    for tick in range(0, ticks):
        values.append(world.get_cost_adjusted_value())
        world.tick()
    values.append(world.get_cost_adjusted_value())
    return values

def run_args_collection(ticks, args_collection):
    results = {}
    for label in args_collection:
        results[label] = run_simulation(ticks, args_collection[label])
    return pd.DataFrame(results)

MAX_TICKS = 13

logging.basicConfig(level=logging.ERROR)

args_collection = {
    #"no_costs": args_no_costs,
    #"bare_minimal": args_minimal,
    "rent": args_rent,
    "apartment": args_apartment,
    "apartment_expensive": args_apartment_expensive,
    "house": args_house,
    "rentvest_apartment": args_rentvest_apartment,
    "rentvest_house": args_rentvest_house,
    "rent_cash": args_rent_cash
}

for resolution_params in [ParameterResolution.PESSIMISTIC, ParameterResolution.MIDDLE, ParameterResolution.OPTIMISTIC]:
    resolved_args_collection = resolve_parameters(args_collection, resolution_params)
    results_table = run_args_collection(MAX_TICKS, resolved_args_collection)
    print(resolution_params)
    print(results_table)
    results_table.plot(figsize=(18,8), grid=True)
```
